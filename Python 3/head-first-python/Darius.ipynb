```python
def parse_filename(filename):
    L = filename.removesuffix(".txt").split('-')
    name, age, distance, stroke = L
    age = int(age)
    stroke = stroke.lower()
    return (name, age, distance, stroke)

fn = "Darius-13-100m-Fly.txt"
r = parse_filename(fn)
print(r)
fn = 'Abi-10-50m-Back.txt'
r = parse_filename(fn)
print(r)
```
